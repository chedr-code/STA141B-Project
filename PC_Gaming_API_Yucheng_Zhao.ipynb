```python
import requests
import time

def search_games(genre, limit=200): 
    base_url = "https://itunes.apple.com/search"
    parameters = {
        "term": genre + " game",
        "media": "software",
        "entity": "software",
        "limit": limit,
        "attribute": "genreIndex",
        "country": "US"
    }

    response = requests.get(base_url, params=parameters)
    if response.status_code == 200:
        return response.json()

def within_time(release_date, start_year=2018, end_year=2023):
    year = int(release_date[0:4])
    return start_year <= year <= end_year

genres = ["Action", "Horror", "Adventure", "Simulation", "Shooter", "Puzzle"]
results = []

for genre in genres:
    games_data = search_games(genre)
    if games_data:
        for game in games_data["results"]:
            release_date = game.get('releaseDate')
            if release_date != 'N/A' and is_within_timeframe(release_date):
                results.append({
                    "Title": game['trackName'],
                    "Genre": genre,
                    "Release Date": release_date,
                    "Average User Rating": game.get('averageUserRating'),
                    "Number of Ratings": game.get('userRatingCount')
                })
    
    # Sleep to avoid hitting the search limit
    time.sleep(1)

    print(results)
```

```python
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors


# Sorting the results by 'Number of Ratings'
results_sorted = sorted(results, key=lambda x: x.get('Number of Ratings', 0), reverse=True)

# Extracting year from release date and preparing data for plotting
release_years = [int(game['Release Date'][0:4]) for game in results_sorted]
ratings = [game.get('Number of Ratings', 0) for game in results_sorted]

# Generating a unique color for each genre
genres = list(set(game['Genre'] for game in results_sorted))
colors = {genre: plt.cm.tab20(i/len(genres)) for i, genre in enumerate(genres)}
plot_colors = [colors[game['Genre']] for game in results_sorted]

# Creating scatter plot
plt.figure(figsize=(12, 8))
plt.scatter(release_years, ratings, c=plot_colors, alpha=0.5)

# Creating a legend
legend_labels = {genre: colors[genre] for genre in genres}
markers = [plt.Line2D([0], [0], marker='o', color=color, label=genre, linestyle='None') for genre, color in legend_labels.items()]
plt.legend(handles=markers, title="Genres")

plt.title('Game Ratings by Release Year and Genre')
plt.xlabel('Release Year')
plt.ylabel('Number of User Ratings')
plt.xlim([2017.5, 2023.5])  # Adjusting x-axis to display years from 2018 to 2023
plt.ylim([min(ratings), max(ratings)])  # Adjusting y-axis based on ratings range
plt.show()

```

```python

ratings_per_year = {}

for game in results:
    year = game['Release Date'][:4]
    genre = game['Genre']
    ratings = game.get('Number of Ratings', 0)
    
    if genre not in ratings_per_year:
        ratings_per_year[genre] = {}
    if year not in ratings_per_year[genre]:
        ratings_per_year[genre][year] = 0
    
    ratings_per_year[genre][year] += ratings


plt.figure(figsize=(12, 8))
# Generate a list of unique colors for the plot
colors = plt.cm.get_cmap('tab10', len(ratings_per_year))

for i, (genre, ratings_per_year) in enumerate(ratings_per_year.items()):
    # Sort the years for consistent plotting
    years = sorted(ratings_per_year.keys())
    total_ratings = [ratings_per_year[year] for year in years]
    plt.plot(years, total_ratings, label=genre, color=colors(i))

plt.title('Total Number of Ratings per Year')
plt.xlabel(years)
plt.ylabel('Number of User Ratings')
plt.legend(title='Genres')
plt.grid(True)
plt.show()


```

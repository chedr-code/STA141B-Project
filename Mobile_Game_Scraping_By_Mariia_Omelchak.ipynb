```python
#Let's see what the spread fo genres looks like in general 
genreratio = finaldf['Genre'].value_counts(normalize = True)
plt.bar(genreratio.index, genreratio.value, color = "pink")
```

```python
#Function for 2018-2023 All game genre ratios (Free + Paid)
fr = []
each_genre = finaldf['Genre'].unique()
genreclr = plt.cm.rainbow(range(len(each_genre)))
each_year = finaldf['Year'].unique()

for year in each_year:
    yeardf = finaldf[finaldf['Year'] == year]
    gen_ratio = yeardf['Genre'].value_counts(normalize = True)
    
    plt.figure(figsize=(14,9))
    sns.barplot( x = gen_ratio.index, y = gen_ratio.values, palette = 'Spectral')
    
    plt.legend(each_genre, loc = 'upper right', bbox_to_anchor = (1,1))
    plt.xlabel("Genre")
    plt.ylabel("Ratio of Games in the Genre")
    plt.title(f'Ratio of IOS Free + Paid Game Genres in {year}')
    
    nameme = f'{year} genres (all).png'
    plt.savefig(nameme, bbox_inches = 'tight')
    fr.append(imageio.imread(nameme))
    
    plt.close()
```
